```python
from selenium import webdriver
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup as bs
from selenium.common.exceptions import WebDriverException
import time
import pandas as pd
import requests
```

```python
chromedriver = "C:/Users/insrik00/Downloads/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(chromedriver)
```

```python
CRN=pd.DataFrame()#columns=['Name','Location','URL','Top Executive','Technology Partners Include:','AWS Level','Cisco Level','HPE Level','IBM Level','NetApp Level','Oracle Level','Microsoft Level','VMware Level','Previous Tech Elite Appearances'])
```

```python
t0=time.clock()
# series_list=[]
for i in range(11,501):
    URL="https://www.crn.com/rankings-and-lists/msp2020-details.htm?c="+str(i)
    print(URL)
    driver.get(URL)
    time.sleep(1.5)
    html_source = driver.page_source
    soup = bs(html_source, 'lxml')
    results = soup.find(id="databaseResults")
    question = results.find_all(class_="quest")
    ans = results.find_all(class_="ans")
    collist=[x.getText() for x in question]
    anslist=[y.getText() for y in ans]
    collist.append('Name')
    anslist.append(soup.find_all('h1')[1].getText())
    #series_list.append(pd.Series(anslist,index=collist))
    CRN=CRN.append(pd.Series(anslist,index=collist),ignore_index=True)
t1 = time.clock() - t0
print(t1)
```

```python
CRN['URL_clean']=CRN['URL'].apply(lambda x: x.replace('http://',''))
```
